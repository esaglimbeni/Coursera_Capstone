## Project Capstone - Toronto's Neighbourhood Analysis

##### This function searches for 'table' elements and only for 'tr' and 'th' rows and 'td' elements within each 'tr' or 'th' element in the table. 'td' stands for “table data”.

```python
import pandas as pd
html_tables= pd.read_html('https://en.m.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df=html_tables[0]
print(df.head())
print(df.tail())
df.shape
```

##### With the drop function, and the condition that the Borough is "Not Assigned", we make the first depuration of the dataframe .

```python
df = df.drop(df[df['Borough']=='Not assigned'].index)
```

```python
df.head(50)
```

##### The next step of this depuration, is about replace the Nighbourhood "Not Assigned" with the same name of the "Borough"

```python
d = df[df['Neighbourhood']=='Not assigned'].index
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df.loc[d].Borough
df.head(50)
```

##### The last process is to ombinate the Neighbourhood's that has the same Postecode - Borough, in the same row.

```python
dh = df.groupby('Postcode').agg(lambda x: ' , '.join(set(x)))#(lambda x: ','.join(x))
dh.reset_index()
dh.head(30)
```
